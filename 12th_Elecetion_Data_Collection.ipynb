#Bangladesh 12th Election Data Analysis

```python
import requests        # Used for web scraping (sending HTTP requests)
from bs4 import BeautifulSoup   # Used for parsing HTML content
```

```python
from google.colab import drive  # Mount Google Drive
drive.mount('/content/drive')
```

###Now, create three folders in Google Drive:
####Bronze – for storing raw data

####Silver – for storing preprocessed data

###Gold – for storing final, analysis-ready data

```python
bronz_folder_path='/content/drive/MyDrive/Data/Bronze'
```

```python
# Define the URL of the 12th National Parliament Election results page
url = 'https://www.ecs.gov.bd/page/result-for-12th-national-parliament-election'

# Send an HTTP GET request to fetch the webpage content
response = requests.get(url)
```

```python
# Parse the HTML content of the response using BeautifulSoup
# 'html.parser' is the parser used to interpret the HTML structure
soup = BeautifulSoup(response.text, 'html.parser')

 #print(soup)
```

```python
fil_links = [] #  to save excel file link of all Seat

```

```python
# Find the first <table> element in the parsed HTML
table = soup.find('table')

# Check if a table exists on the page
if table:
  # Find all rows (<tr>) inside the table
  rows = table.find_all('tr')

  # Loop through each row
  for row in rows:
    # Find all cells (<td>) within the row
    cells = row.find_all('td')

    # Loop through each cell to look for links
    for cell in cells:
       link = cell.find('a')  # Find <a> tag (hyperlink) inside the cell

       # If a link exists, get its 'href' attribute
       if link:
        href = link.get('href')

        # Check if the link points to an Excel file (.xlsx)
        if '.xlsx' in href:
          fil_links.append(href)  # Add the Excel file link to the list
```

```python
#for link in fil_links: # view all files save correctly
  #print(link)
```

```python
# Loop through each Excel file link collected earlier
for link in fil_links:
   # Extract the file name from the URL (text after the last '/')
# Example from the link https://www.ecs.gov.bd/bec/public/files/1/12th-parliament-election-constituency-wise-result/NPE_24_1.xlsx
# file_name is "NPE_24_1.xlsx"

   file_name = link.split('/')[-1]

   # Send an HTTP GET request to download the Excel file
   response = requests.get(link)

   # Define the full path where the file will be saved
   link_to_Save_file = bronz_folder_path +"/"+ file_name

   # Open the file in write-binary mode and save the downloaded content
   with open(link_to_Save_file, 'wb') as file:
      file.write(response.content)
```
